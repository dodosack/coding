```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------- ACHSEN EINSTELLEN ----------------

# Achsengrenzen — BEI ANDEREN PUNKTEN ODER FUNKTION ANPASSEN!
x_min, x_max = 0.0, 6.0
y_min, y_max = -1.0, 2.0

# Zeichenfläche (Größe abhängig vom Achsbereich)
fig, ax = plt.subplots(figsize=((x_max - x_min), (y_max - y_min)))
ax.axis([x_min, x_max, y_min, y_max])
display(fig)

# Rasterpunkte (Hauptachsen-Ticks)
x_ticks = np.arange(x_min, x_max + 1, 1)
print('x_ticks = ', x_ticks)
ax.set_xticks(x_ticks)

y_ticks = np.arange(y_min, y_max + 1, 1)
print('y_ticks = ', y_ticks)
ax.set_yticks(y_ticks)
display(fig)

# Gitter anzeigen
ax.grid(True)
display(fig)

# Nebenraster aktivieren
ax.set_xticks(np.arange(x_min, x_max + 1), minor=True)
ax.set_yticks(np.arange(y_min, y_max + 1), minor=True)
ax.grid(which='minor', alpha=0.1)
ax.grid(which='major', alpha=0.5)
display(fig)

# ---------------- PUNKTE PLOTTEN ----------------

# Die 5 Stützpunkte — BEI NEUER AUFGABE HIER ÄNDERN!
ax.plot(1, 0, color='red', marker='*', markersize=5)
ax.plot(2, 0, color='red', marker='*', markersize=5)
ax.plot(4, 0, color='red', marker='*', markersize=5)
ax.plot(5, 0, color='red', marker='*', markersize=5)
ax.plot(3, 1, color='red', marker='*', markersize=5)
display(fig)

# ---------------- POLYNOM AUFSTELLEN ----------------

# X- und Y-Koordinaten der Stützpunkte — HIER PUNKTE EINTRAGEN!
x_kurve = np.array([1, 2, 3, 4, 5])
y_kurve = np.array([0, 0, 1, 0, 0])

# Erstelle das Gleichungssystem (Matrix B) für 4. Ordnung (n=5 Punkte)
B = np.array([
    [x_kurve[0] ** 4, x_kurve[0] ** 3, x_kurve[0] ** 2, x_kurve[0], 1],
    [x_kurve[1] ** 4, x_kurve[1] ** 3, x_kurve[1] ** 2, x_kurve[1], 1],
    [x_kurve[2] ** 4, x_kurve[2] ** 3, x_kurve[2] ** 2, x_kurve[2], 1],
    [x_kurve[3] ** 4, x_kurve[3] ** 3, x_kurve[3] ** 2, x_kurve[3], 1],
    [x_kurve[4] ** 4, x_kurve[4] ** 3, x_kurve[4] ** 2, x_kurve[4], 1]
])

# Löse das Gleichungssystem B * a = y → Finde die Koeffizienten a
koeffizienten = np.linalg.solve(B, y_kurve)

# ---------------- POLYNOM DEFINIEREN UND PLOTTEN ----------------

# Definiere das berechnete Polynom
def polynom(x):
    return (
        koeffizienten[0] * x ** 4 +
        koeffizienten[1] * x ** 3 +
        koeffizienten[2] * x ** 2 +
        koeffizienten[3] * x +
        koeffizienten[4]
    )

# Erzeuge viele x-Werte zum Zeichnen der Kurve
x_werte = np.linspace(0, 6, 500)
y_werte = polynom(x_werte)

# Zeichne die Kurve
ax.plot(x_werte, y_werte, color='Blue')
display(fig)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# ---------------- PARAMETER ----------------

# Radius des rollenden Kreises – HIER ÄNDERN FÜR ANDERE ZYKLOIDEN
r = 2

# Parameter t: Bereich der Kurve – 4π = 2 volle Umdrehungen bei r=2
# Wenn du z.B. nur eine Umdrehung brauchst, nutze t = np.linspace(0, 2 * np.pi, 500)
t = np.linspace(0, 4 * np.pi, 1000)

# ---------------- ZYKLOIDE BERECHNEN ----------------

# Parametrisierung der Zykloide:
# x(t) = r*(t - sin(t)), y(t) = r*(1 - cos(t))
x = r * (t - np.sin(t))
y = r * (1 - np.cos(t))

# ---------------- PLOT ERSTELLEN ----------------

fig, ax = plt.subplots(figsize=(10, 4))  # Plotgröße

# Zeichne die Kurve
ax.plot(x, y, color='black', linewidth=2)

# ---------------- ACHSENBESCHRIFTUNG ----------------

ax.set_xlabel("x")  # X-Achse beschriften
ax.set_ylabel("y")  # Y-Achse beschriften

# ---------------- ANSICHT EINSTELLEN ----------------

# Plotbereich festlegen – BEI ANDEREN RÄDERN/T-Werten ANPASSEN
ax.set_xlim(-1, 26)
ax.set_ylim(-1, 5)

# Haupt-Ticks (Gitterlinien bei Ganzzahlen)
ax.set_xticks(np.arange(-1, 27, 1))
ax.set_yticks(np.arange(-1, 6, 1))

# Neben-Ticks bei 0.5er-Schritten (feines Raster)
ax.xaxis.set_minor_locator(MultipleLocator(0.5))
ax.yaxis.set_minor_locator(MultipleLocator(0.5))

# ---------------- GITTERLINIEN ----------------

# Gitterlinien für Haupt- und Nebenachsen
ax.grid(which='major', color='lightgray', linewidth=0.7)
ax.grid(which='minor', color='lightgray', linewidth=0.3)

# Gleiche Skalierung auf beiden Achsen (Winkel korrekt sichtbar)
ax.set_aspect('equal')

# ---------------- ANZEIGE ----------------
plt.show()
```
