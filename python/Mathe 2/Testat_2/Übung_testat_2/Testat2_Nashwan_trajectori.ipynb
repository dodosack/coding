```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------- ACHSEN-EINSTELLUNGEN ----------------

# Begrenzung der x- und y-Achsen (diese Werte kannst du bei Bedarf anpassen)
x_min, x_max = -5.0, 5.0  # X-Achse von -5 bis +5
y_min, y_max = -5.0, 5.0  # Y-Achse von -5 bis +5

# Größe der Zeichenfläche automatisch an den Achsenbereich angepasst
fig, ax = plt.subplots(figsize=((x_max - x_min) / 4, (y_max - y_min) / 4))
ax.set_aspect('equal')  # Gleiche Skalierung auf beiden Achsen
ax.axis([x_min, x_max, y_min, y_max])  # Achsenbereich festlegen

# ---------------- RASTER UND TICKS ----------------

# Positionen der x-Ticks (Rasterpunkte auf der x-Achse) — bei Bedarf ändern
x_ticks = np.arange(x_min, x_max + 1, 1)
print('x_ticks = ', x_ticks)
ax.set_xticks(x_ticks)

# Positionen der y-Ticks (Rasterpunkte auf der y-Achse)
y_ticks = np.arange(y_min, y_max + 1, 1)
print('y_ticks = ', y_ticks)
ax.set_yticks(y_ticks)

# Hauptgitter anzeigen
ax.grid(True)
display(fig)

# Nebenraster hinzufügen (für feinere Orientierung)
ax.set_xticks(np.arange(x_min, x_max + 1), minor=True)
ax.set_yticks(np.arange(y_min, y_max + 1), minor=True)
ax.grid(which='minor', alpha=0.1)  # Nebenraster leicht sichtbar
ax.grid(which='major', alpha=0.5)  # Hauptraster etwas kräftiger
display(fig)

# ---------------- KURVE DEFINIEREN ----------------

# Wertebereich für den Parameter t (Steuerung der Kurve)
# Wenn du z.B. nur eine Umdrehung willst: reduziere den Wert von 8π
t = np.linspace(0, 8 * np.pi, 1000)

# x- und y-Koordinaten der Kurve — ÄNDERE DIESE FUNKTION BEI ANDEREN KURVEN!
# Aktuell: Spirale mit √t * cos(t) und √t * sin(t)
x = np.sqrt(t) * np.cos(t)  # X-Funktion der Kurve
y = np.sqrt(t) * np.sin(t)  # Y-Funktion der Kurve

# Zeichne die Kurve (schwarz)
ax.plot(x, y, color='black')
display(fig)
```
