```python
import numpy as np
import matplotlib.pyplot as plt

# Gegebene Punkte
x = np.array([1, 2, 3, 4, 5])
y = np.array([0, 0, 1, 0, 0])

# Vandermonde-Matrix und Lösung des linearen Gleichungssystems
V = np.vander(x, increasing=False)
coeffs = np.linalg.solve(V, y)

# Polynomfunktion
def poly(x_val):
    return sum(c * x_val**i for i, c in enumerate(reversed(coeffs)))

# Plotwerte
x_plot = np.linspace(0, 6.5, 400)
y_plot = poly(x_plot)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_plot, y_plot, 'b-', label='Polynom')
plt.plot(x, y, 'r*', markersize=10, label='Gegebene Punkte')

# Gitter mit festen Schritten

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(np.arange(0, 6.6, 1))  # x-Achse in 1er-Schritten
plt.yticks(np.arange(-1, 2.6, 0.5))  # y-Achse in 0.5er-Schritten

# Achsengrenzen und Achsenverhältnis
plt.xlim(0, 6.5)
plt.ylim(-1, 2.5)
plt.axis('equal')

# Achsenbeschriftung und Titel
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolierendes Polynom 4. Grades durch 5 Punkte')
plt.legend()
plt.tight_layout()
plt.show()

# Koeffizienten anzeigen
for i, c in enumerate(reversed(coeffs)):
    print(f'a{i} = {c}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_min, x_max, y_min, y_max = 0.0, 6.0, -1.0, 2.0

fig, ax = plt.subplots(figsize=((x_max - x_min), (y_max - y_min)))
ax.axis([x_min, x_max, y_min, y_max])
display(fig)
x_ticks = np.arange(x_min, x_max + 1, 1)
print('x_ticks = ', x_ticks)
ax.set_xticks(x_ticks)
y_ticks = np.arange(y_min, y_max + 1, 1)
print('y_ticks = ', y_ticks)
ax.set_yticks(y_ticks)
display(fig)
ax.grid(True)
display(fig)
ax.set_xticks(np.arange(x_min, x_max + 1), minor=True)
ax.set_yticks(np.arange(y_min, y_max + 1), minor=True)
ax.grid(which='minor', alpha=0.1)
ax.grid(which='major', alpha=0.5)
display(fig)
ax.plot(1, 0, color='red', marker='*', markersize=5)
ax.plot(2, 0, color='red', marker='*', markersize=5)
ax.plot(4, 0, color='red', marker='*', markersize=5)
ax.plot(5, 0, color='red', marker='*', markersize=5)
ax.plot(3, 1, color='red', marker='*', markersize=5)
display(fig)
x_kurve = np.array([1, 2, 3, 4, 5])
y_kurve = np.array([0, 0, 1, 0, 0])

# Erstelle die Matrix für das Polynom der 4. Ordnung
B = np.array([
    [x_kurve[0] ** 4, x_kurve[0] ** 3, x_kurve[0] ** 2, x_kurve[0], 1],
    [x_kurve[1] ** 4, x_kurve[1] ** 3, x_kurve[1] ** 2, x_kurve[1], 1],
    [x_kurve[2] ** 4, x_kurve[2] ** 3, x_kurve[2] ** 2, x_kurve[2], 1],
    [x_kurve[3] ** 4, x_kurve[3] ** 3, x_kurve[3] ** 2, x_kurve[3], 1],
    [x_kurve[4] ** 4, x_kurve[4] ** 3, x_kurve[4] ** 2, x_kurve[4], 1]
])

# Löse das lineare Gleichungssystem für die Koeffizienten des Polynoms
koeffizienten = np.linalg.solve(B, y_kurve)


# Definiere das Polynom mithilfe der berechneten Koeffizienten
def polynom(x):
    return koeffizienten[0] * x ** 4 + koeffizienten[1] * x ** 3 + koeffizienten[2] * x ** 2 + koeffizienten[3] * x + \
        koeffizienten[4]


# Plotten
x_werte = np.linspace(0, 6, 500)
y_werte = polynom(x_werte)
ax.plot(x_werte, y_werte, color='Blue')
display(fig)
```
